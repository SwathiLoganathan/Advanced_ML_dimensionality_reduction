```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset1=pd.read_csv("prep.csv",index_col=None)

df2=dataset1

df = pd.get_dummies(df2, drop_first=True)

X=df.drop('classification_yes',axis=1)
Y=df['classification_yes']
```

```python
df
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# Applying LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components = 1)
X_train1 = lda.fit_transform(X_train,y_train)
X_test1 = lda.transform(X_test)

explained_variance = lda.explained_variance_ratio_
feature_names=lda.get_feature_names_out(input_features=None)

print("lda:",lda)
print("explained_variance:",explained_variance)
print("feature_names:",feature_names)
```

```python
explained_variance
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train1, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test1)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
        
Accuracy=accuracy_score(y_test, y_pred)
report=classification_report(y_test, y_pred)
Accuracy
```
